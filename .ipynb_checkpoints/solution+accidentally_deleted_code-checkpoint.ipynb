```python
### DATA CLEANSING
```

```python
## Removing duplicate values and Creating a unique film identifier
```

```python
data_set_1 = pd.read_csv("IMDb movies.csv") # Raw Data to confirm that .drop_duplicates() is working without inplace=True
```

```python
data_set_1.info()
```

```python
data_set_1 = pd.read_csv("IMDb movies.csv").drop_duplicates(subset="title") # Removing duplicate values within the 1st data set
data_set_1["title_year"] = data_set_1["title"] + "-" + data_set_1["year"].astype(str) # Unique film identifier within the 1st data set
```

```python
data_set_1.head(3)
```

```python
data_set_1.info() # By using .drop_duplicates(), record number went down to 76618 from 81273
```

```python
data_set_2 = pd.read_csv("the_oscar_award.csv").drop_duplicates(subset="film") # Removing duplicate values within the 2nd data set
data_set_2["title_year"] = data_set_2["film"] + "-" + data_set_2["year_film"].astype(str) # Unique film identifier within the 2nd data set
```

```python
data_set_2.head(3)
```

```python
data_set_2.info()
```

```python
## Merging 2 .csv files by a common column using a Left Join with Pandas
```

```python
data_set = pd.merge(data_set_1, data_set_2, how="left") # Merging two data sets
```

```python
## Dropping columns that are not necessary to the analysis
```

```python
data_set.drop(columns=["imdb_title_id", "original_title", "language", "year", "metascore", "actors", "description", "worlwide_gross_income", "reviews_from_users", "reviews_from_critics", "ceremony", "year_film", "film", "year_ceremony"], inplace = True)
```

```python
data_set.head(3)
```

```python
data_set.info()
```

```python
## Removing records, which do not include day and month of the film published date
```

```python
data_set.drop(data_set[data_set["date_published"].apply(lambda x: len(str(x)) < 10)].index, inplace=True)
```

```python
data_set["date_published_parced"] = pd.to_datetime(data_set["date_published"])
data_set.drop(columns=["date_published"], inplace=True)
```

```python
## Removing any Non Numeric characters and Spaces from both "budget" and "usa_gross_income" series
```

```python
data_set["budget"] = data_set["budget"].fillna("$ 0").apply(lambda x: x.split(" ")[1])
```

```python
data_set["usa_gross_income"] = data_set["usa_gross_income"].fillna("$ 0").apply(lambda x: x.split(" ")[1])
```

```python
## Analysing and eliminating the NaN values across the remaining Series
```

```python
data_set[["country", "director", "writer", "production_company"]] = data_set[["country", "director", "writer", "production_company"]].fillna("Not Specified")
```

```python
data_set[["category", "name", "winner"]] = data_set[["category", "name", "winner"]].fillna("Not Nominated for an Oscar")
```

```python
data_set.head(3)
```

```python
data_set.info()
```

```python
## Changing "budget" and "usa_gross_income" data type to integer
```

```python
data_set["budget"] = data_set["budget"].apply(lambda x: int(x))
```

```python
data_set["usa_gross_income"] = data_set["usa_gross_income"].apply(lambda x: int(x))
```

```python
data_set.info()
```

```python
## Find out when it is potentially the best time to Release a Film based on the Release date
```

```python
data_set["day_published"] = data_set["date_published_parced"].dt.dayofweek
```

```python
data_set["day_published"] = data_set["day_published"].replace({0: "Monday", 1: "Tuesday", 2: "Wednesday", 3: "Thursday", 4: "Friday", 5: "Saturday", 6: "Sunday"})
```

```python
data_set["winner"] = data_set["winner"].replace({False: "Did not win", True: "Won an Oscar"}) # changing values
```

```python
### DATA VISUALIZATION
```

```python
## Find out which Film Genres are most successful based both on Average Viewers vote as well as the Won Oscars
```

```python
top_genres_based_on_viewer_votes = data_set.groupby("genre")[["avg_vote", "votes", "winner"]].max().round().sort_values(["avg_vote"], ascending = False)[0:20]
```

```python
top_genres_based_on_viewer_votes["genre_name"] = top_genres_based_on_viewer_votes.index # creating an extra column to contain values of the index
```

```python
top_genres_based_on_viewer_votes
```

```python
plt.subplots(figsize = (14, 7)) # Comparing Averafe Votes based on the total number of the votes made
ax = sns.barplot(x = "genre_name", y = "avg_vote", data = top_genres_based_on_viewer_votes)
ax.set(xlabel = "Genres", ylabel = "Average Vote")
ax2 = ax.twinx()
ax.set_title("Top Genres based on User Votes")
ax2.set_ylabel("Total Number of Votes", rotation = 90)
top_genres_based_on_viewer_votes.plot(x="genre_name", y="votes", ax=ax2, legend=False, color="r")
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45);
```

```python
##  Find out which Film Genres are best to invest in at present based on Budget and Income
```

```python
top_genres_to_invest_in = data_set.groupby("genre")[["budget", "usa_net_income"]].max().round().sort_values(["usa_net_income"], ascending = False)[0:20]
```

```python
top_genres_to_invest_in["genre_name"] = top_genres_to_invest_in.index
```

```python
plt.subplots(figsize = (20, 7)) # Comparing Budget invested and Income made
ax = sns.lineplot(x = "genre_name", y = "budget", data = top_genres_to_invest_in)
ax.set(xlabel = "Genres", ylabel = "Production Budget")
ax2 = ax.twinx()
ax.set_title("Top Genres to Invest In")
ax2.set_ylabel("USA NET Income", rotation = 90)
top_genres_to_invest_in.plot(x="genre_name", y="usa_net_income", ax=ax2, legend=True, color="r")
plt.setp(ax.get_xticklabels(), rotation = 45);
```
